```python
# IMPORTAR A BIBLIOTECA PANDAS
import pandas as pd
```

```python
# LER O ARQUIVO
DF = pd.read_csv("/content/drive/MyDrive/GSPC.csv")
# MOSTAR O CABEÇALHO E AS 5 PRIMEIRAS LINHAS
DF.head()
```

```python
# DELETAR A COLUNA 'Date'
DF = DF.drop('Date',axis=1)
```

```python
# MOSTRAR A TABELA NOVAMENTE
DF.head()
```

```python
# VISUALIZAR AS ÚLTIMAS DUAS LINHAS DA BASE DE DADOS
DF[-2::]
```

```python
# CRIAR UMA VARIÁVEL CHAMADA 'amanha' E VOU QUERER A ÚLTIMA LINHA DA BASE DE DADOS
amanha = DF [-1::]
amanha
```

```python
# CRIAR UMA VARIÁVEL QUE RECEBE TODA A BASE DE DADOS EXCETO A ÚLTIMA LINHA
base = DF.drop(DF[-1::].index, axis=0)
```

```python
# MOSTRAR O CABEÇALHO E AS ÚLTIMAS 5 LINHAS
base.tail()
```

```python
# CRIAR UMA NOVA COLUNA ELA VAI RECEBER O CONTEÚDO DA COLUNA 'Close' PORÉM VAI COMEÇAR A PARTIR DA LINHA 1 E NÃO DA LINHA 0
base['target'] = base['Close'][1:len(base)].reset_index(drop=True)

# MOSTRAR O CABEÇALHO E AS ÚLTIMAS 5 LINHAS
base.tail()
```

```python
# OBS: DA NOVA COLUNA CRIADA A ÚLTIMA LINHA 'NaN' VAI SER USADA PARA A PREVISÃO DO MODELO
```

```python
# CRIAR UMA VARIÁVEL QUE DA COLUNA 'target' VAI EXCLUIR A ÚLTIMA LINHA, OU SEJA A QUE ESTA VAZIA
prev = base[-1::].drop('target', axis = 1)

prev
```

```python
# CRIAR OUTRA VARIÁVEL QUE AGORA VAI EXCLUIR TODA A LINHA ANTERIOR '17216'
Treino = base.drop(base[-1::].index,axis = 0)
Treino.tail()
```

```python
# TRANSFORMAR OS DAOS DA COLUNA 'target' EM 0 OU 1 

# DA BASE DE DADOS 'treino' SI O VALOR DA LINHA 'target' FOR MAIOR QUE O DA linha 'CLOSE' ENTÃO COLOQUE 1
Treino.loc[Treino['target'] > Treino['Close'],'target'] = 1
Treino.tail()
```

```python
# AGORA SI NA COLUNA 'target' O VALOR FOR DIFERENTE DE 1 ENTÃO COLOQUE 0
Treino.loc[Treino['target'] != 1, 'target'] = 0
Treino.tail()
```

PREPARAR OS DADOS PARA UTILIZAR O MODELO DE MACHINE LEARNING

```python
# SEPARAR A VARIÁVEL X DEPENDENTE E A VARIÁVEL Y INDEPENDENTE
x = Treino.drop('target', axis = 1)
y = Treino['target']
```

```python
# SEPARAR UMA PARTE DA BASE DE DADOS PARA TREINO E OUTRA PARA TESTE 
from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)
```

```python
# IMPORTAR O MODELO DE MACHINE LEARNING CLASSIFICADOR DE ÁRVORES EXTRAS e TREINAR A INTELIGÊNCIA ARTIFICIAL
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)
```

```python
# MOSTRAR A PONTUAÇÃO DO TESTE
Resultado = modelo.score(x_teste, y_teste)
print('Acurácia: ',Resultado)
```

```python
# VISUALIZAR A VARIÁVEL 'prev'
prev
```

```python
# O MODELO VAI MOSTRAR O VALOR DA VARIÁVEL SI É 1 OU 0
modelo.predict(prev)
```

```python
# COLOCAR A VARIÁVEL 'prev' DENTRO DA MINHA COLUNA 'target'
prev['target'] = modelo.predict(prev)
prev
```

```python
# CÓDIGO PARA RETIRAR ATIVOS
def warn(*args, **kwargs):
    pass
import warnings
warnings.wain = warn

# IMPORTAR E LER O ARQUIVO
import pandas as pd
prev = pd.read_csv('/content/drive/MyDrive/prev.csv')
# O CÓDIGO VAI MOSTRAR O VALOR DO FECHAMENTO ANTERIOR E DA PREVISÃO ANTERIOR
print('Fechamento anterior: ', prev['Close'][0])
print('Previsão anterior: ',prev['target'][0])

# DEPOIS VAI SER MOSTRAR O VALOR DE HOJE
base = pd.read_csv('/content/drive/MyDrive/Hoje.csv')


try:
    amanha = pd.read_csv('/content/drive/MyDrive/Futuro.csv')
    print('Fechamento atual: ', amanha['Close'][0])
    base = base.append(amanha[:1],sort = True)
    amanha = amanha.drop(amanha[:1].index,axis = 0)
    base.to_csv('/content/drive/MyDrive/Hoje.csv',index = False)
    amanha.to_csv('/content/drive/MyDrive/Futuro.csv',index = False)
except Exception:
    print('O fechamento ainda não ocorreu!')
    pass

base['target'] = base['Close'][1:len(base)].reset_index(drop=True)
prev = base[-1::].drop('target', axis = 1)
Treino = base.drop(base[-1::].index,axis = 0)
Treino.loc[Treino['target'] > Treino['Close'],'target'] = 1
Treino.loc[Treino['target'] != 1, 'target'] = 0
x = Treino.drop('target', axis = 1)
y = Treino['target']

from sklearn.model_selection import train_test_split
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y, test_size=0.3)
from sklearn.ensemble import ExtraTreesClassifier
modelo = ExtraTreesClassifier()
modelo.fit(x_treino, y_treino)

print('Acurácia: ',modelo.score(x_teste, y_teste))

prev['target'] = modelo.predict(prev)
print('Fechamento de ontem: ',prev['Close'][0])

if prev ['target'][0]==1:
  print('VAI SUBIR!')
else:
  print('VAI CAIR!')

prev.to_csv('/content/drive/MyDrive/prev.csv',index = False)
```
